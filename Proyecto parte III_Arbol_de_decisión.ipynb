## variable Fase_dia

```python
tabla_IV = df.groupby('Fase_dia').agg(      
    Cant = ('C_vitimas', 'count'),
    Cant_vitimas = ('C_vitimas', 'sum'),    
    Tasa_vitimas = ('C_vitimas', 'mean')
).reset_index()
tabla_IV['Tasa_SV'] = 1 - tabla_IV['Tasa_vitimas']
tabla_IV['Cant_SV'] = tabla_IV['Cant'] - tabla_IV['Cant_vitimas']
tabla_IV['LN'] = np.log(tabla_IV['Tasa_vitimas'] / tabla_IV['Tasa_SV'])
tabla_IV['Pct_Vitimas'] = tabla_IV['Cant_vitimas'] / sum(tabla_IV['Cant_vitimas'])
tabla_IV['Pct_SV'] = tabla_IV['Cant_SV'] / sum(tabla_IV['Cant_SV'])
tabla_IV['Resta'] = tabla_IV['Pct_Vitimas'] - tabla_IV['Pct_SV']
tabla_IV['IV'] = tabla_IV['Resta'] * tabla_IV['LN']
tabla_IV
```

## Árbol de decisión

Comencé seleccionando las variables 'Fase_dia' y 'Sentido_via' como predictores para mi modelo de árbol de decisiones. Estas variables las escogí por su potencial predictivo que identifiqué mediante el análisis de Valor de Información (IV).
La variable 'Fase_dia' resultó tener el mayor IV, seguida por 'Sentido_via', lo que me indica que ambas son variables importantes para predecir la cantidad de víctimas en accidentes de tráfico.

Convertí las variables categóricas 'Fase_dia' y 'Sentido_via' en numéricas mediante codificación, para que pudieran ser utilizadas por mi modelo. Además, creé la matriz de predictores X y el vector objetivo y para los siguientes pasos.

```python
# Supongamos que 'Fase_dia' es una columna categórica que contiene 'Pleno dia', 'Plena Noite', 'Anoitecer' y 'Amanhecer'
df['F_dia'] = df['Fase_dia'].map({'Pleno dia': 1, 'Plena Noite': 2, 'Anoitecer': 3, 'Amanhecer': 0})

# Corregir el nombre de la columna 'Descent_Grouped'
# Supongamos que 'Descent_Grouped' es una columna categórica que contiene 'Decrescente', 'Crescente' y 'Não Informado'
df['S_via'] = df['Sentido_via'].map({'Decrescente': 1, 'Crescente': 2, 'Não Informado': 0})

# Imprimir el DataFrame con las nuevas columnas
print(df[['F_dia', 'S_via']])
```

La transformación fue exitosa y resultó en una matriz X con datos de entrada listos para el modelo, y un vector y que contiene la variable objetivo de mi estudio, el número de víctimas ('C_vitimas').

```python
# Enlisto los predictores

lista_predictores = ['F_dia', 'S_via']
```

```python
# Matriz predictores

X = df[lista_predictores].values
```

```python
# Vector de target

y = df['C_vitimas'].values
```

```python

from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.tree import plot_tree
```

```python
#Genero modelo

clf = DecisionTreeClassifier(random_state = 1234, # Semilla aleatoria
                             criterion = 'gini', # Funcion para medir calidad de la división
                             splitter = 'best', # Estrategia para elegir la división de cada nodo
                             max_depth = 4, # Profundidad máxima del árbol
                             min_samples_leaf = 20 # Mínima cantidad de registros por nodo final
                            )
clf.feature_names = lista_predictores
model = clf.fit(X, y)
```

Configuré y entrené un modelo de árbol de decisiones (DecisionTreeClassifier) con parámetros específicos diseñados para controlar la complejidad del árbol y prevenir el sobreajuste.
El modelo se entrenó sin problemas y quedó listo para realizar predicciones. Establecí una semilla aleatoria para asegurar la reproducibilidad del modelo.

```python
plt.figure(figsize=(12, 8))
plot_tree(clf, 
          filled = True, 
          rounded = True, 
          class_names = ['F_dia','S_via'], 
          feature_names = lista_predictores, 
          impurity = False,
          proportion = True,
          node_ids = True)
```

Visualización del Árbol de Decisiones: Después del entrenamiento, visualicé el árbol utilizando plot_tree, configurando la visualización para que fuera fácilmente interpretable y revelara las proporciones de las clases en cada nodo del árbol.

La visualización del árbol me permitió entender cómo el modelo utiliza 'Fase_dia' y 'Sentido_via' para hacer predicciones. Observé que la variable S_via es utilizada como el primer criterio de división, lo que indica su fuerte influencia en la clasificación final.

Examiné las descripciones de los nodos generados por el modelo, que me proporcionaron detalles sobre las decisiones en cada punto del árbol.

Descubrí que ciertas condiciones de 'Fase_dia' y 'Sentido_via' se asocian con mayores o menores números de víctimas. Esto fue evidente por las distribuciones de clases en los nodos hoja y por la clase predominante en cada uno.

```python
# Genero predicciones  
df['Predict_Arbol_Clase'] = clf.predict(df[lista_predictores])
df['Predict_Arbol_Nodo'] = clf.apply(df[lista_predictores])
```
