### Pregunta 1: ¿Cuál es la tendencia en el número total de accidentes a lo largo de los años?

En 2020, el gráfico comienza con aproximadamente 63,400 accidentes.
Luego, en 2021, hay un pico significativo que muestra un aumento hasta alrededor de 64,200 accidentes.
Para 2022, se observa una disminución pronunciada hasta aproximadamente 63,600 accidentes.
El pico en 2021 puede sugerir varios factores, como cambios en las condiciones de la carretera, modificaciones en las leyes de tráfico, o incluso factores externos como el clima o eventos especiales que podrían haber contribuido al aumento de los accidentes.

```python
accidentes_por_año = df['Año'].value_counts().sort_index().reset_index()
accidentes_por_año.columns = ['Año', 'Número de acidentes']


plt.figure(figsize=(12, 6))
plt.plot(accidentes_por_año['Año'], accidentes_por_año['Número de acidentes'], marker='o', linestyle='-')
plt.title('Tendência no número total de acidentes por ano')
plt.xlabel('Año')
plt.ylabel('Número de acidentes')
plt.grid(True)

plt.xticks(accidentes_por_año['Año'], rotation=45)


resumen_estadistico = accidentes_por_año.describe()

print(resumen_estadistico)

plt.tight_layout()
plt.show()
```

## Pregunta 2: ¿Existe una estacionalidad en la ocurrencia de accidentes a lo largo de los meses o las estaciones del año?

Este análisis permite analizar y visualizar la distribución de accidentes a lo largo de los meses y las estaciones del año, proporcionando dos gráficos de barras y estadísticas resumidas para comprender mejor la variación de accidentes en función del tiempo

Analizando el primer gráfico, parece haber una variación menor en la cantidad de accidentes de tráfico a lo largo de los meses, con los números fluctuando alrededor de una media que parece estar entre 12,000 y 15,000 accidentes por mes. No hay una tendencia clara o picos significativos que sugieran una estacionalidad fuerte en los datos mensuales.

En el segundo gráfico, que muestra la distribución por estación del año, la frecuencia de accidentes también se mantiene relativamente constante entre las estaciones, aunque hay una leve disminución durante la primavera. Los números son aproximadamente:

Otoño: Cerca de 50,000 accidentes
Invierno: Ligeramente por debajo de 50,000 accidentes
Verano: Algo más de 40,000 accidentes
Primavera: Alrededor de 40,000 accidentes


```python
df['Mes'] = df['Año_acidente'].dt.month
df['Estación'] = df['Mes'].apply(lambda x: 'Invierno' if (x >= 1 and x <= 2) or x == 12 else
                                            'Primavera' if x >= 3 and x <= 5 else
                                            'Verano' if x >= 6 and x <= 8 else 'Otoño')


accidentes_por_mes = df['Mes'].value_counts().sort_index().reset_index()
accidentes_por_mes.columns = ['Mes', 'Número de acidentes']


accidentes_por_estacion = df['Estación'].value_counts().reset_index()
accidentes_por_estacion.columns = ['Estación', 'Número de acidentes']


plt.figure(figsize=(12, 6))


plt.subplot(1, 2, 1)
plt.bar(accidentes_por_mes['Mes'], accidentes_por_mes['Número de acidentes'])
plt.title('Distribuição de acidentes por mês')
plt.xlabel('Mes')
plt.ylabel('')
plt.grid(True)


plt.subplot(1, 2, 2)
plt.bar(accidentes_por_estacion['Estación'], accidentes_por_estacion['Número de acidentes'])
plt.title('Distribuição de acidentes por estação do ano')
plt.xlabel('Estación del Año')
plt.ylabel('')
plt.grid(True)


resumen_mes = accidentes_por_mes.describe()
resumen_estacion = accidentes_por_estacion.describe()


print("Resumen por Mes:")
print(resumen_mes)

print("\nResumen por Estación del Año:")
print(resumen_estacion)

plt.tight_layout()
plt.show()
```

##   3. ¿Cuáles son las condiciones meteorológicas más comunes en los accidentes y cómo afectan a la gravedad de los mismos "Mortos, Feridos_leves, Feridos_graves, Ilesos o Ignorados"?

Muestran una visualización de barras apiladas que representa la frecuencia de las condiciones meteorológicas en función de la gravedad de los accidentes y proporciona estadísticas descriptivas para cada categoría de gravedad en relación con las condiciones meteorológicas. Esto ayuda a entender cómo las condiciones meteorológicas pueden influir en la gravedad de los accidentes.

Cielo Claro (Céu Claro): Es la condición más comúnmente reportada en los accidentes durante los tres años, con una frecuencia que supera los 30,000 eventos anualmente.
Lluvia (Chuva): Es la segunda condición más frecuente, con una disminución notable en el año 2022 comparado con los años anteriores.
Sol (Sol) y Viento (Vento): Estas condiciones aparecen con menos frecuencia en los reportes de accidentes.
Ignorado: Representa los casos en los que las condiciones meteorológicas no fueron reportadas o desconocidas. La frecuencia de esta categoría es significativamente baja comparada con las condiciones de Cielo Claro y Lluvia.
Condiciones menos comunes: Granizo (Granizo), Niebla/Neblina (Nevoeiro/Nebulina), y Nieve (Neve) tienen las frecuencias más bajas en los datos. Esto puede ser debido a que estas condiciones son menos comunes en las áreas estudiadas o que los conductores toman precauciones adicionales bajo estas condiciones.

Las condiciones menos comunes como la nieve y la neblina, a pesar de su baja frecuencia, podrían ser de interés para estudiar más a fondo debido a su potencial para causar accidentes graves. La categoría Ignorado destaca la importancia de una buena recopilación de datos para asegurar la calidad y la utilidad de los análisis de seguridad vial

```python
warnings.filterwarnings("ignore")

columns_of_interest = ['Condicao_metereologica', 'Año', 'Mortos', 'Feridos_leves', 'Feridos_graves', 'Ilesos', 'Ignorados']
filtered_df = df[columns_of_interest]

frequency_table = filtered_df.groupby(['Condicao_metereologica', 'Año']).size().reset_index(name='Frecuencia')


frequency_table['Condicao_metereologica'] = frequency_table['Condicao_metereologica'].astype(pd.CategoricalDtype())
summary_statistics = frequency_table.groupby('Condicao_metereologica', observed=True)['Frecuencia'].describe()

plt.figure(figsize=(12, 8))
sns.set_palette("viridis")


sns.barplot(data=frequency_table, x='Condicao_metereologica', y='Frecuencia', hue='Año', estimator=sum, errorbar=None, order=frequency_table['Condicao_metereologica'].unique())


print(summary_statistics)

plt.title('Frequência das condições meteorológicas por ano')
plt.xlabel('Condições climatéricas')
plt.ylabel('')
plt.xticks(rotation=45)
plt.tight_layout()
plt.legend(title='Año')
plt.show()

warnings.filterwarnings("default")
```

 ##  4. ¿Qué días de la semana tienen la mayor cantidad de accidentes y cómo varía esta distribución?
Permite comparar la frecuencia de accidentes en diferentes días de la semana a lo largo de los años entre 2020 a 2022, lo que puede ayudar a identificar patrones o tendencias en la ocurrencia de accidentes en función del día de la semana y el año.

Los accidentes son menos frecuentes al comienzo de la semana, con los lunes (segunda-feira) mostrando las cifras más bajas.
Hay un marcado aumento hacia el final de la semana, con los viernes (sexta-feira) y sábados (sábado) registrando los números más altos de accidentes.
Esta tendencia es consistente a lo largo de los tres años, aunque las cifras absolutas pueden variar de un año a otro.
La frecuencia de accidentes aumenta significativamente durante los fines de semana, lo que podría estar relacionado con un aumento de actividades recreativas y viajes durante estos días.
2022 muestra una disminución general en la frecuencia de accidentes en comparación con los años anteriores, lo que podría indicar una posible mejora en las condiciones de seguridad vial o cambios en los patrones de viaje, tal vez influenciados por eventos externos o políticas públicas.


```python
accidents_by_day_and_year = df.groupby(['Año', 'Dia_semana']).size().unstack()

order = ['domingo', 'segunda-feira', 'terça-feira', 'quarta-feira', 'quinta-feira', 'sexta-feira', 'sábado']
accidents_by_day_and_year = accidents_by_day_and_year[order]

plt.figure(figsize=(12, 8))
sns.set_palette("husl")


accidents_by_day_and_year_transposed = accidents_by_day_and_year.T


for year in accidents_by_day_and_year_transposed.columns:
    plt.plot(accidents_by_day_and_year_transposed.index, accidents_by_day_and_year_transposed[year], marker='o', label=str(year))

plt.title('Frequência de acidentes por dia da semana')
plt.xlabel('')
plt.ylabel('')
plt.xticks(rotation=45)
plt.tight_layout()
plt.legend(title='Año')
plt.show()

summary_statistics = accidents_by_day_and_year.describe()
print("Resumen de Accidentes por Día de la Semana:")
print(summary_statistics)
```
